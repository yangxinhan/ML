# MinMaxScaler

## 載入套件

```python
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

# 設定中文字型
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']  
# 矯正負號
plt.rcParams['axes.unicode_minus'] = False
```

## 載入資料

```python
# 常態分配
data = np.random.normal(10, 5, 10000)
plt.figure(figsize=(5,3));
plt.hist(data, 20)
plt.show()
```

## 轉換

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_new =scaler.fit_transform(data.reshape(-1, 1))
plt.figure(figsize=(5,3));
plt.hist(data_new, 20)
plt.show()
```

## 均勻分配轉換

```python
# 均勻分配
data = np.random.uniform(0, 10, 10000)
plt.figure(figsize=(5,3));
plt.hist(data, 20)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_new =scaler.fit_transform(data.reshape(-1, 1))
plt.figure(figsize=(5,3));
plt.hist(data_new, 20)
plt.show()
```

## Gamma分配轉換

```python
# Gamma 分配
data = np.random.gamma(1, 2, 10000)
plt.figure(figsize=(5,3));
plt.hist(data, 20)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_new =scaler.fit_transform(data.reshape(-1, 1))
plt.figure(figsize=(5,3));
plt.hist(data_new, 20)
plt.show()
```

## 簡單測試

```python
data = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]
scaler = MinMaxScaler()
scaler.fit(data)
print(scaler.data_min_, scaler.data_max_)
```

```python
data_new = scaler.transform(data)
print(data_new)
```

```python
# 會超出範圍
data2 = [[-2, 20]]
data_new = scaler.transform(data2)
print(data_new)
```

## 資料還原

```python
scaler.inverse_transform(data_new)
```

## 設定參數 feature_range，改變轉換後的範圍(-1, 1)

```python
scaler = MinMaxScaler(feature_range=(-1, 1))
data_new = scaler.fit_transform(data)
print(data_new)
```

## 設定參數 clip=True，強制裁剪，轉換後的資料不會超出範圍

```python
scaler = MinMaxScaler(clip=True, feature_range=(-1, 1))
data_new = scaler.fit_transform(data)
print(data_new)
```

```python
# 不會超出範圍
data2 = [[-2, 20]]
data_new = scaler.fit_transform(data2)
print(data_new)
```
