# 臨界點與優化(Critical Points and Optimization)

## 繪圖函數

```python
from matplotlib import pyplot as plt

def create_plot(f):
    x = list(range(0, 11))

    # Use the function to get the y values
    y = [f(i) for i in x]

    # Set up the graph
    plt.xticks(range(0, 15, 1))
    plt.yticks(range(-200, 500, 20))
    plt.xlabel('x')
    plt.ylabel('k(x)')
    plt.grid()

    # Plot the function
    plt.plot(x,y, color='green')
```

$\begin{equation}k(x) = -10x^{2} + 100x + 3 \end{equation}$

```python
# 測試 
k = lambda x: -10*(x**2) + (100*x)  + 3
create_plot(k)
```

## 求導數
### $\begin{equation}k'(x) = -20x + 100 \end{equation}$

```python
# 測試 
kd = lambda x: -20*x + 100
create_plot(k)

# 導數
x = list(range(0, 11))
yd = [kd(i) for i in x]
# 繪圖
plt.plot(x,yd, color='purple')
plt.show()
```

## 函數加負號

```python
def create_plot2(f):
    x = list(range(0, 11))

    # Use the function to get the y values
    y = [f(i) for i in x]

    # Set up the graph
    plt.xticks(range(0, 15, 1))
    plt.yticks(range(-300, 120, 20))
    plt.xlabel('x')
    plt.ylabel('k(x)')
    plt.grid()

    # Plot the function
    plt.plot(x,y, color='green')
    
def k(x):
    return -(-10*(x**2) + (100*x)  + 3)
def kd(x):
    return -(-20*x + 100)
create_plot2(k)

# 導數
x = list(range(0, 11))
yd = [kd(i) for i in x]
# 繪圖
plt.plot(x, yd, color='purple')
plt.ylim(-300, 120)
plt.show()
```

## 下圖變化率從正值逐漸轉為負值，即斜率=0時，y有最大值

```python
# 測試 
k = lambda x: -10*(x**2) + (100*x)  + 3
kd = lambda x: -20*x + 100
create_plot(k)

# 導數
x = list(range(0, 11))
yd = [kd(i) for i in x]
# 繪圖
plt.plot(x,yd, color='purple')

# Plot tangent slopes for x = 2, 5, and 8
x1 = 2
x2 = 5
x3 = 8
plt.plot([x1-1,x1+1],[k(x1)-(kd(x1)),k(x1)+(kd(x1))], color='red')
plt.plot([x2-1,x2+1],[k(x2)-(kd(x2)),k(x2)+(kd(x2))], color='red')
plt.plot([x3-1,x3+1],[k(x3)-(kd(x3)),k(x3)+(kd(x3))], color='red')

plt.show()
```

## 臨界點(Critical Points)
### 導數=0常常是最大值或最小值，但未必一定是，如下。
### $\begin{equation}v(x) = x^{3} - 2x + 100 \end{equation}$

```python
# Create function v
def v(x):
    return (x**3) - (2*x) + 100

def vd(x):
    return 3*(x**2) - 2

def create_plot3(f):
    x = list(range(-10, 11))

    # Use the function to get the y values
    y = [f(i) for i in x]

    # Set up the graph
    plt.xticks(range(-10, 15, 1))
    plt.yticks(range(-1000, 2000, 100))
    plt.xlabel('x')
    plt.ylabel('k(x)')
    plt.grid()

    # Plot the function
    plt.plot(x,y, color='green')
    
create_plot3(v)
# 導數
x = list(range(-10, 11))
yd = [vd(i) for i in x]
# 繪圖
plt.plot(x,yd, color='purple');
```

### 紫色線是綠色曲線的導數，綠色曲線是一個馬鞍線(saddle)，斜率由正值趨近於0，又由0遞增為正值
### 導數為0的點，稱為臨界點(critical point)
### 拐點(inflexion point)：導數由正轉負，形成一個峰(peak)，是一個區域最大值(local maximum)，反之，導數由負轉正，形成一個槽(trough)，是一個區域最小值(local minimum)。

## 求最大值或最小值
### $\begin{equation}k(x) = -10x^{2} + 100x + 3 \end{equation}$

### $\begin{equation}k'(x) = -20x + 100 \end{equation}$

### $\begin{equation}-20x + 100 = 0 \end{equation}$

### $\begin{equation}x = 5 \end{equation}$

## 只知 x = 5 可能是最大值、最小值或只是一個拐點(inflexion)
## 進一步觀察二階導數(Second Order Derivatives)
## 二階導數
### $\begin{equation}k'(x) = -20x + 100 \end{equation}$
### $\begin{equation}k''(x) = -20 \end{equation}$

###  若二階導數是常數，可以推斷一階導數是線性函數，若常數<0，表一階導數(變化率)逐步遞減，所以，越過一階導數=0的點，就會逐步遞減，因此可以推斷 一階導數=0 的點有最大值。

```python
import matplotlib

# Create function k
def k(x):
    return -10*(x**2) + (100*x)  + 3

def kd(x):
    return -20*x + 100

def k2d(x):
    return -20

from matplotlib import pyplot as plt

# Create an array of x values to plot
x = list(range(0, 11))

# Use the function to get the y values
y = [k(i) for i in x]

# Use the derivative function to get the k'(x) values
yd = [kd(i) for i in x]

# Use the 2-derivative function to get the k''(x)
y2d = [k2d(i) for i in x]

# 修改中文字體
zhfont1 = matplotlib.font_manager.FontProperties(fname='C:\Windows\Fonts\msjh.ttc', size=10)

# Set up the graph
plt.xlabel('x (time in seconds)')
plt.ylabel('k(x) (height in feet)',fontproperties=zhfont1)
plt.xticks(range(0,15, 1))
plt.yticks(range(-200, 500, 20))
plt.grid()

# Plot the function
plt.plot(x,y, color='green')

# Plot  k'(x)
plt.plot(x,yd, color='purple')

# Plot k''(x)
plt.plot(x,y2d, color='blue')

plt.show()
```

### 反之，若二階導數是常數，若常數>0，表一階導數逐步遞增，所以，越過一階導數=0的點，就會逐步遞增，因此可以推斷 一階導數=0的點有最小值。

## 臨界點不是最大值、最小值或只是一個拐點(inflexion)
### $\begin{equation}v(x) = x^{3} - 6x^{2} - 12x + 2 \end{equation}$
### $\begin{equation}v'(x) = 3x^{2} - 12x - 12 \end{equation}$
### $\begin{equation}3x^{2} - 12x - 12 = 0 \end{equation}$
### $\begin{equation}3x(x - 4) = 12 \end{equation}$
### $\begin{equation}x(x - 4) = 4 \end{equation}$
### $\begin{equation}x^{2} - 4x = 4 \end{equation}$
### $\begin{equation}x^{2} - 4x + 4 = 0 \end{equation}$
### $\begin{equation}(x - 2)^{2} = 0 \end{equation}$
### $\begin{equation}x - 2 = \pm\sqrt{0}\end{equation}$
### $\begin{equation}x - 2 = +\sqrt{0} = 0, -\sqrt{0} = 0\end{equation}$
### v'(2) = 0
### $\begin{equation}v''(x) = 6x - 12\end{equation}$
### $\begin{equation}v''(2) = 0\end{equation}$

```python
# Create function v
def v(x):
    return (x**3) - (6*(x**2)) + (12*x) + 2

def vd(x):
    return (3*(x**2)) - (12*x) + 12

def v2d(x):
    return (3*(2*x)) - 12

# Create an array of x values to plot
x = list(range(-5, 11))

# Use the function to get the y values
y = [v(i) for i in x]

# Use the derivative function to get the derivative values
yd = [vd(i) for i in x]

# Use the derivative function to get the derivative values
y2d = [v2d(i) for i in x]

# Set up the graph
plt.xlabel('x')
plt.ylabel('v(x)')
plt.xticks(range(-10,15, 1))
plt.yticks(range(-2000, 2000, 50))
plt.grid()

# Plot the function
plt.plot(x,y, color='green')

# Plot the derivative
plt.plot(x,yd, color='purple')

# Plot the derivative
plt.plot(x,y2d, color='blue')

plt.show()
```
