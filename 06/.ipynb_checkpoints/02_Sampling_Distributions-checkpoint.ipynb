# 抽樣(Sampling)與抽樣分配(Sampling Distributions)
## 母體(population)：研究對象全體所構成之集合。
## 樣本(sample)：從母體中選取一些代表性的子集合。
## 普查(census)：對所要研究的某種現象作全面性的調查。
## 抽樣(Sampling)：對所要研究的某種現象作部分的調查。
### 資料來源：[國立高雄大學統計學研究所網頁](https://www.stat.nuk.edu.tw/prost/content2/statics_1.htm)

### 抽樣：進行田野調查，通常因為母體(Population)的資料量過大，無法進行普查(census)，只能抽取一部份樣本，以推測母體的行為模式。
### 例如要針對2022年台北市長選舉進行一場民意調查，母體為台北市民且年齡>=20歲，進行抽樣1000份，以推測各候選人當選的機率。

## 中央極限定理(Central Limit Theorem)
### 不論任何分配的樣本，如果樣本批數夠大，每批樣本的平均值都會近似常態分配(normal distribution)。

## 驗證：抽取 10,000批，每批 100個乘客的樣本

```python
import pandas as pd

# https://numpy.org/doc/stable/reference/random/generated/numpy.random.binomial.html
n, p, s = 100, 0.25, 10000
df = pd.DataFrame(np.random.binomial(n,p,s)/n, columns=['p-hat'])

# Plot the distribution as a histogram
means = df['p-hat']
means.plot.hist(title='Simulated Sampling Distribution')  
plt.show()
print ('Mean: ' + str(means.mean()))
print ('Std: ' + str(means.std()))
```

## 抽樣分配的平均數與標準誤差(Mean and Standard Error of a Sampling Distribution of Proportion)
### 樣本分配是基於各樣本的平均數，而非所有樣本，它的標準差稱為標準誤差(standard error)，公式如下：

### $\begin{equation}\sigma_{\hat{p}} = \sqrt{\frac{p(1-p)}{n}}\end{equation}$
## 從同學中隨機抽樣16位，調查抽菸的比例，標準差計算如下：
### $\begin{equation}\sigma_{\hat{p}} = \sqrt{\frac{0.25 \times 0.75}{16}} \approx 0.11\end{equation}$

## 反覆實驗100批，標準誤差(Standard Error)計算如下:
### $\begin{equation}\sigma_{\hat{p}} = \sqrt{\frac{0.25 \times 0.75}{100}} \approx 0.043\end{equation}$

## 當樣本數愈多時，標準誤差會隨之減少，表示預測變異會愈小.

## 信賴區間(confidence interval)：在某一信心水準下所作的區間估計，例如2倍標準差下，信賴區間=[0.164, 0.336]。
## 計算如下：
## 抽樣誤差(margin of error, MoE)：在某一信心水準下，產生的標準誤差，例如2倍標準差，MoE = 2 x 0.043&plusmn;  0.086。
### 信賴區間 = 0.25 &plusmn; 0.086 = [0.164, 0.336].
## 顯著水準(significance level) = 1 - 信心水準，常用5%。

## [標準誤差(standard error)證明](https://en.wikipedia.org/wiki/Standard_error)

## [信賴區間(confidence interval) 定義](https://zh.wikipedia.org/zh-tw/信賴區間)

### 95%信賴區間代表有95%的信心，抽樣的資料會落在此區間內。
### 95%信賴區間不代表有95%的樣本資料落在此信賴區間。

## 信賴區間繪圖

```python
n, p, s = 100, 0.25, 10000
df = pd.DataFrame(np.random.binomial(n,p,s)/n, columns=['p-hat'])

# Plot the distribution as a histogram
means = df['p-hat']
m = means.mean()
sd = means.std()
moe1 = m - (sd * 2)
moe2 = m + (sd * 2)


means.plot.hist(title='Simulated Sampling Distribution')  

plt.axvline(m, color='red', linestyle='dashed', linewidth=2)
plt.axvline(moe1, color='magenta', linestyle='dashed', linewidth=2)
plt.axvline(moe2, color='magenta', linestyle='dashed', linewidth=2)
plt.show()
```

## 使用 norm.interval 函數計算常態分配的信賴區間。

```python
from scipy import stats

mu, sigma, n = 3.2, 1.2, 500
samples = list(range(0, 1000))

# data will hold all of the sample data
data = np.array([])

# sampling will hold the means of the samples
sampling = np.array([])

# Perform 1,000 samples
for s in samples:
    # In each sample, get 500 data points from a normal distribution
    sample = np.random.normal(mu, sigma, n)
    data = np.append(data,sample)
    sampling = np.append(sampling,sample.mean())

# Create a dataframe with the sampling of means
df = pd.DataFrame(sampling, columns=['mean'])

# Get the Mean, StdDev, and 95% CI of the means
means = df['mean']
m = means.mean()
sd = means.std()
ci = stats.norm.interval(0.95, m, sd)

# Plot the distribution, mean, and CI
means.plot.hist(title='Simulated Sampling Distribution', bins=100) 
plt.axvline(m, color='red', linestyle='dashed', linewidth=2)
plt.axvline(ci[0], color='magenta', linestyle='dashed', linewidth=2)
plt.axvline(ci[1], color='magenta', linestyle='dashed', linewidth=2)
plt.show()

# Print the Mean, StdDev and 95% CI
print ('Sampling Mean: ' + str(m))
print ('Sampling StdErr: ' + str(sd))
print ('95% Confidence Interval: ' + str(ci))
```

```python
# 信賴區間驗算
m - 1.96 * sd, m + 1.96 * sd
```
