```python
import pandas as pd
import yfinance as yt
from datetime import datetime
import mplfinance as mpf
```

```python
# 取得歷史股價
today = datetime.today().strftime('%Y-%m-%d')
df = yt.download('2317.TW', start='2016-01-01', end=today)
df.tail()
```

```python
mpf.plot(df[-30:])
```

```python
mpf.plot(df[-30:], type='candle')
```

```python
mpf.plot(df, type='line')
```

```python
mpf.plot(df, type='pnf')
```

```python
mpf.plot(df[-300:], type='candle', mav=(5, 20, 60))
```

```python
mpf.plot(df[-300:], type='candle', mav=(5, 20, 60), volume=True)
```

## Style
https://github.com/matplotlib/mplfinance/blob/master/examples/styles.ipynb

```python
mpf.plot(df[-30:], type='candle', style='charles')
```

```python
mc = mpf.make_marketcolors(up='g',down='r')
s  = mpf.make_mpf_style(marketcolors=mc)
mpf.plot(df[-30:], type='candle', style=s)
```
