```python
import pandas as pd
import yfinance as yt
import datetime
import mplfinance as mpf
```

```python
def add_days_to_date_string(date_string1, days=1):
    date2 = datetime.datetime.strptime(date_string1, r"%Y-%m-%d").date() + datetime.timedelta(days=days)
    return  date2.strftime(r"%Y-%m-%d")
```

```python
# 取得歷史股價
end_date = datetime.datetime.today().strftime('%Y-%m-%d')
start_date = add_days_to_date_string(end_date, -120)
df = yt.download('2303.TW', start=start_date, end=end_date)
df.tail()
```

```python
mpf.plot(df[-30:])
```

```python
mpf.plot(df[-30:], type='candle')
```

```python
mpf.plot(df, type='line')
```

```python
mpf.plot(df, type='pnf')
```

```python
mpf.plot(df[-300:], type='candle', mav=(5, 20, 60))
```

```python
mpf.plot(df[-300:], type='candle', mav=(5, 20, 60), volume=True)
```

## Style
https://github.com/matplotlib/mplfinance/blob/master/examples/styles.ipynb

```python
mpf.plot(df[-30:], type='candle', style='charles')
```

```python
mc = mpf.make_marketcolors(up='g',down='r')
s  = mpf.make_mpf_style(marketcolors=mc)
mpf.plot(df[-30:], type='candle', style=s)
```
